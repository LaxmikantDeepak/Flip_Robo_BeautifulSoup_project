```python
from bs4 import BeautifulSoup

import requests
```

```python
page=requests.get("https://bookpage.com/")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
book1=soup.find_all('h4',class_='italic')
book1
```

```python
books=[]
for i in book1:
    books.append(i.text.replace('\n',''))
    
books
```

```python
b_auther=soup.find_all('p',class_='sans bold')
b_auther
```

```python
authers=[]
for i in b_auther:
    authers.append(i.text.replace('\n',''))
    
authers
```

```python
genre=soup.find_all('p',class_='genre-links')
genre
```

```python
genre_all=[]
for i in genre:
    genre_all.append(i.text.replace('\n',''))
    
genre_all
```

```python
b_review=soup.find_all('p',class_='excerpt')
b_review
```

```python
all_review=[]
for i in b_review:
    all_review.append(i.text.replace('<p class="excerpt">',''))
    
all_review
```

```python
all_review=[]
for i in b_review:
    all_review.append(i.text.replace('\n',''))
    
all_review
```

```python
import pandas as pd
pd.set_option("display.max_colwidth", None)
all_books=pd.DataFrame({})
```

```python
all_books['Books_name']=books
all_books['Auther']=authers
all_books['Genre']=genre_all
all_books['Review']=all_review
```

```python
books=all_books[:5]
```

```python
books.to_csv( "all_books.csv", index=False, encoding='utf-8-sig')
```
